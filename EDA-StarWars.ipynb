### Explore the Demographics of the Respondents

Let's start by exploring the age, gender, and location of the respondents. We'll visualize these demographics to better understand the composition of our dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Create subplots for Age, Gender, and Location
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))

# Plot Age distribution
sns.countplot(data=df, x='Age', ax=axes[0])
axes[0].set_title('Age Distribution')
axes[0].set_xlabel('Age Groups')
axes[0].set_ylabel('Count')

# Plot Gender distribution
sns.countplot(data=df, x='Gender', ax=axes[1])
axes[1].set_title('Gender Distribution')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Count')

# Plot Location distribution
sns.countplot(data=df, y='Location (Census Region)', ax=axes[2])
axes[2].set_title('Location Distribution')
axes[2].set_xlabel('Count')
axes[2].set_ylabel('Location (Census Region)')

plt.tight_layout()
plt.show()
```

Here are some insights from the demographic distributions:

    - Age Distribution: The age group "45-60" has the highest number of respondents, followed by the "30-44" age group. The "18-29" and "60+" age groups are less represented in the dataset.

    - Gender Distribution: The dataset has more male respondents compared to female respondents.

    - Location Distribution: Respondents are primarily from the "East North Central" and "Pacific" regions, with other regions being less represented.

### 4.2 Analyze the Popularity of the Star Wars Movies

Next, let's find out which Star Wars movies are the most and least popular among the respondents. We'll look at the number of people who have watched each movie and their rankings.

```python
# Columns related to which movies the respondents have seen
movies_seen_columns = df.columns[3:9]

# Rename these columns for easier analysis
new_column_names = [
    'Seen Episode I',
    'Seen Episode II',
    'Seen Episode III',
    'Seen Episode IV',
    'Seen Episode V',
    'Seen Episode VI'
]

# Create a mapping dictionary and rename the columns
rename_dict = {old: new for old, new in zip(movies_seen_columns, new_column_names)}
df.rename(columns=rename_dict, inplace=True)

# Convert the values to 'Yes' if the respondent has seen the movie, and 'No' otherwise
for col in new_column_names:
    df[col] = df[col].apply(lambda x: 'Yes' if pd.notna(x) else 'No')

# Count the number of 'Yes' for each movie
movie_popularity = df[new_column_names].apply(lambda x: x == 'Yes').sum()

# Plot the popularity of each movie
plt.figure(figsize=(12, 6))
sns.barplot(x=movie_popularity.index, y=movie_popularity.values)
plt.title('Popularity of Star Wars Movies')
plt.xlabel('Movies')
plt.ylabel('Number of Views')
plt.show()
```

The bar chart indicates the number of respondents who have seen each Star Wars episode:

    - Seen Episode V: This appears to be the most popular movie, with the highest number of views among the respondents.
    - Seen Episode III: This movie has the least number of views, making it the least popular among the respondents.

### 4.1 Univariate Analysis

    - Distribution of respondents by gender
    - Distribution of respondents by age

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove the first row which contains 'Response' as it seems to be metadata
df = df.iloc[1:]

# Convert some of the columns to appropriate types for analysis
df['RespondentID'] = df['RespondentID'].astype('str')

# Dropping rows where gender or age is NaN for the purpose of visualization
df = df.dropna(subset=['Gender', 'Age'])

# Set up the matplotlib figure
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Univariate Analysis
# 1. Distribution of respondents by Gender
sns.countplot(x='Gender', data=df, ax=axes[0])
axes[0].set_title('Distribution of Respondents by Gender')

# 2. Distribution of respondents by Age
sns.countplot(x='Age', data=df, ax=axes[1], order=df['Age'].value_counts().index)
axes[1].set_title('Distribution of Respondents by Age')
axes[1].tick_params(axis='x', rotation=45)

# 3. Distribution of respondents by Household Income
sns.countplot(x='Household Income', data=df, ax=axes[2], order=df['Household Income'].value_counts().index)
axes[2].set_title('Distribution of Respondents by Household Income')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### 4.2 Bivariate Analysis

    - Correlation between age and household income
    - Gender vs. Star Wars film fandom
    - Age vs. Star Wars film fandom

```python
# Bivariate Analysis
# Prepare the data by replacing string-based categorical values with numerical ones for certain columns
# This is done for easier correlation analysis

# Mapping age groups to numerical values
age_mapping = {
    '18-29': 1,
    '30-44': 2,
    '45-60': 3,
    '> 60': 4
}
df['Age_numeric'] = df['Age'].map(age_mapping)

# Mapping household income to numerical values
income_mapping = {
    '$0 - $24,999': 1,
    '$25,000 - $49,999': 2,
    '$50,000 - $99,999': 3,
    '$100,000 - $149,999': 4,
    '$150,000+': 5
}
df['Income_numeric'] = df['Household Income'].map(income_mapping)

# Set up the matplotlib figure
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 1. Correlation between Age and Household Income
sns.scatterplot(x='Age_numeric', y='Income_numeric', data=df, ax=axes[0])
axes[0].set_title('Correlation between Age and Household Income')
axes[0].set_xlabel('Age Group (Numerically Mapped)')
axes[0].set_ylabel('Household Income (Numerically Mapped)')

# 2. Gender vs Star Wars film fandom
sns.countplot(x='Gender', hue='Do you consider yourself to be a fan of the Star Wars film franchise?', data=df, ax=axes[1])
axes[1].set_title('Gender vs Star Wars Film Fandom')

# 3. Age vs Star Wars film fandom
sns.countplot(x='Age', hue='Do you consider yourself to be a fan of the Star Wars film franchise?', data=df, ax=axes[2])
axes[2].set_title('Age vs Star Wars Film Fandom')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

**Correlation between Age and Household Income**

At first glance, there doesn't seem to be a strong correlation between age and household income based on the scatter plot. The data points are spread across the chart without any clear trend. Note that the age and income groups were numerically mapped for easier visualization.


**Gender vs Star Wars Film Fandom**

Among males and females, the majority consider themselves to be fans of the Star Wars franchise. However, the number of male fans appears to be significantly higher than female fans.


**Age vs Star Wars Film Fandom**

Across various age groups, the majority consider themselves fans of the Star Wars franchise. The age group "45-60" has the highest number of fans, followed by "30-44".



### 4.3 Multivariate Analysis

    - Pairplot or heatmap to show correlations between multiple variables, if applicable

```python
# Multivariate Analysis

# Select only the columns that are relevant for the multivariate analysis
# For demonstration purposes, I've selected gender, age, household income, and fandom status
selected_columns = ['Gender', 'Age_numeric', 'Income_numeric', 'Do you consider yourself to be a fan of the Star Wars film franchise?']
df_selected = df[selected_columns].dropna()

# Map 'Yes' and 'No' in fandom status to 1 and 0 respectively for easier correlation analysis
df_selected['Fandom_numeric'] = df_selected['Do you consider yourself to be a fan of the Star Wars film franchise?'].map({'Yes': 1, 'No': 0})

# Map 'Male' and 'Female' in gender to 1 and 0 respectively for easier correlation analysis
df_selected['Gender_numeric'] = df_selected['Gender'].map({'Male': 1, 'Female': 0})

# Calculate the correlation matrix
correlation_matrix = df_selected[['Gender_numeric', 'Age_numeric', 'Income_numeric', 'Fandom_numeric']].corr()

# Draw the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Selected Variables")
plt.show()

# Pairplot to visualize relationships between numerical variables
sns.pairplot(df_selected, hue='Do you consider yourself to be a fan of the Star Wars film franchise?', diag_kind='kde')
plt.suptitle('Pairplot of Selected Variables', y=1.02)
plt.show()
```

**Correlation Matrix of Selected Variables**
    
    -The correlation matrix shows that there is no strong correlation between the selected variables (Gender, Age, Household Income, and Star Wars Film Fandom). The highest correlation value is approximately 0.16, indicating a very weak correlation.


**Pairplot of Selected Variables**

    - The pairplot provides a visualization of the relationships between the selected numerical variables. Different colors represent whether the respondent is a fan of the Star Wars franchise or not.
    - The KDE (Kernel Density Estimation) plots along the diagonal show the distribution of each variable.
    - No clear patterns or trends are visible in the scatter plots, reinforcing the findings from the correlation matrix that there are no strong correlations between these variables.
