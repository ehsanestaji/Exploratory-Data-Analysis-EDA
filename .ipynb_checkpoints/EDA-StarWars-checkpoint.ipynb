```python
# Check for missing values in each column
missing_values = df.isna().sum()

# Check the data types of each column
data_types = df.dtypes

missing_values, data_types
```

    <h3 style="color:#4CAF50; font-family: Arial;">üîÑ 3.3 Data Cleaning Steps</h3>

        We'll convert the entries in the newly renamed <code>Seen_X</code> columns to Boolean values to indicate whether a respondent has seen a particular <strong>Star Wars</strong> movie. This will also handle the missing values by converting them to <code>False</code>.

    <h3 style="color:#4CAF50; font-family: Arial;">üë• Explore the Demographics of the Respondents</h3>

        Let's start by exploring the age, gender, and location of the respondents. We'll visualize these demographics to better understand the composition of our dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Create subplots for Age, Gender, and Location
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))

# Plot Age distribution
sns.countplot(data=df, x='Age', ax=axes[0])
axes[0].set_title('Age Distribution')
axes[0].set_xlabel('Age Groups')
axes[0].set_ylabel('Count')

# Plot Gender distribution
sns.countplot(data=df, x='Gender', ax=axes[1])
axes[1].set_title('Gender Distribution')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Count')

# Plot Location distribution
sns.countplot(data=df, y='Location (Census Region)', ax=axes[2])
axes[2].set_title('Location Distribution')
axes[2].set_xlabel('Count')
axes[2].set_ylabel('Location (Census Region)')

plt.tight_layout()
plt.show()
```

<div style="background-color:#f9f9f9; padding:20px; border-radius:10px; margin-bottom: 20px;">
    <p style="text-indent:2em; font-family: Arial;">
        Here are some insights from the demographic distributions:

    <ul style="font-family: Arial;">
        <li><strong>Age Distribution:</strong> The age group "45-60" has the highest number of respondents, followed by the "30-44" age group. The "18-29" and "60+" age groups are less represented in the dataset.</li>
        <li><strong>Gender Distribution:</strong> The dataset has more male respondents compared to female respondents.</li>
        <li><strong>Location Distribution:</strong> Respondents are primarily from the "East North Central" and "Pacific" regions, with other regions being less represented.</li>
    </ul>
    <hr>
    <h3 style="color:#4CAF50; font-family: Arial;">üé¨ 4.2 Analyze the Popularity of the Star Wars Movies</h3>
    <p style="text-indent:2em; font-family: Arial;">
        Next, let's find out which Star Wars movies are the most and least popular among the respondents. We'll look at the number of people who have watched each movie and their rankings.
    </p>
</div>


```python
# Columns related to which movies the respondents have seen
movies_seen_columns = df.columns[3:9]

# Rename these columns for easier analysis
new_column_names = [
    'Seen Episode I',
    'Seen Episode II',
    'Seen Episode III',
    'Seen Episode IV',
    'Seen Episode V',
    'Seen Episode VI'
]

# Create a mapping dictionary and rename the columns
rename_dict = {old: new for old, new in zip(movies_seen_columns, new_column_names)}
df.rename(columns=rename_dict, inplace=True)

# Convert the values to 'Yes' if the respondent has seen the movie, and 'No' otherwise
for col in new_column_names:
    df[col] = df[col].apply(lambda x: 'Yes' if pd.notna(x) else 'No')

# Count the number of 'Yes' for each movie
movie_popularity = df[new_column_names].apply(lambda x: x == 'Yes').sum()

# Plot the popularity of each movie
plt.figure(figsize=(12, 6))
sns.barplot(x=movie_popularity.index, y=movie_popularity.values)
plt.title('Popularity of Star Wars Movies')
plt.xlabel('Movies')
plt.ylabel('Number of Views')
plt.show()
```

<div style="background-color:#f9f9f9; padding:20px; border-radius:10px; margin-bottom: 20px;">
    <p style="text-indent:2em; font-family: Arial;">
        The bar chart indicates the number of respondents who have seen each Star Wars episode:
    </p>
    <ul style="font-family: Arial;">
        <li><strong>Seen Episode V:</strong> This appears to be the most popular movie, with the highest number of views among the respondents.</li>
        <li><strong>Seen Episode III:</strong> This movie has the least number of views, making it the least popular among the respondents.</li>
    </ul>

    <h3 style="color:#4CAF50; font-family: Arial;">üìä 4.1 Univariate Analysis</h3>
    <ul style="font-family: Arial;">
        <li><strong>Distribution of respondents by gender</strong></li>
        <li><strong>Distribution of respondents by age</strong></li>
        <li><strong>Distribution of respondents by household income</strong></li>
    </ul>
</div>


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove the first row which contains 'Response' as it seems to be metadata
df = df.iloc[1:]

# Convert some of the columns to appropriate types for analysis
df['RespondentID'] = df['RespondentID'].astype('str')

# Dropping rows where gender or age is NaN for the purpose of visualization
df = df.dropna(subset=['Gender', 'Age'])

# Set up the matplotlib figure
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Univariate Analysis
# 1. Distribution of respondents by Gender
sns.countplot(x='Gender', data=df, ax=axes[0])
axes[0].set_title('Distribution of Respondents by Gender')

# 2. Distribution of respondents by Age
sns.countplot(x='Age', data=df, ax=axes[1], order=df['Age'].value_counts().index)
axes[1].set_title('Distribution of Respondents by Age')
axes[1].tick_params(axis='x', rotation=45)

# 3. Distribution of respondents by Household Income
sns.countplot(x='Household Income', data=df, ax=axes[2], order=df['Household Income'].value_counts().index)
axes[2].set_title('Distribution of Respondents by Household Income')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

<h3 style="color: #2E86C1; font-family: Arial, sans-serif;">üìä <strong>Distribution of Respondents by Gender</strong></h3>
<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">The majority of respondents are male. This could be indicative of the Star Wars franchise's primary audience, or it might be related to the sample of people who took this survey.</p>

<hr>

<h3 style="color: #2E86C1; font-family: Arial, sans-serif;">üìä <strong>Distribution of Respondents by Age</strong></h3>
<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">The age group most represented in the dataset is "45-60", followed by "30-44". The least represented age group is "60+". This could provide insights into which age groups are most interested in the Star Wars franchise.</p>

<hr>

<h3 style="color: #2E86C1; font-family: Arial, sans-serif;">üìä <strong>Distribution of Respondents by Household Income</strong></h3>
<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">Most respondents did not provide information about their household income, as indicated by the large number of missing values. Among those who did, the most common income range is "$50,000 - $99,999".</p>

<hr>

<h3 style="color: #34495E; font-family: Arial, sans-serif;">üìà <strong>4.2 Bivariate Analysis</strong></h3>
<ul style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">
    <li>Correlation between age and household income</li>
    <li>Gender vs. Star Wars film fandom</li>
    <li>Age vs. Star Wars film fandom</li>
</ul>


```python
# Bivariate Analysis
# Prepare the data by replacing string-based categorical values with numerical ones for certain columns
# This is done for easier correlation analysis

# Mapping age groups to numerical values
age_mapping = {
    '18-29': 1,
    '30-44': 2,
    '45-60': 3,
    '> 60': 4
}
df['Age_numeric'] = df['Age'].map(age_mapping)

# Mapping household income to numerical values
income_mapping = {
    '$0 - $24,999': 1,
    '$25,000 - $49,999': 2,
    '$50,000 - $99,999': 3,
    '$100,000 - $149,999': 4,
    '$150,000+': 5
}
df['Income_numeric'] = df['Household Income'].map(income_mapping)

# Set up the matplotlib figure
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 1. Correlation between Age and Household Income
sns.scatterplot(x='Age_numeric', y='Income_numeric', data=df, ax=axes[0])
axes[0].set_title('Correlation between Age and Household Income')
axes[0].set_xlabel('Age Group (Numerically Mapped)')
axes[0].set_ylabel('Household Income (Numerically Mapped)')

# 2. Gender vs Star Wars film fandom
sns.countplot(x='Gender', hue='Do you consider yourself to be a fan of the Star Wars film franchise?', data=df, ax=axes[1])
axes[1].set_title('Gender vs Star Wars Film Fandom')

# 3. Age vs Star Wars film fandom
sns.countplot(x='Age', hue='Do you consider yourself to be a fan of the Star Wars film franchise?', data=df, ax=axes[2])
axes[2].set_title('Age vs Star Wars Film Fandom')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

<h3 style="color: #2E86C1; font-family: Arial, sans-serif;">üìà <strong>Correlation between Age and Household Income</strong></h3>
<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">At first glance, there doesn't seem to be a strong correlation between age and household income based on the scatter plot. The data points are spread across the chart without any clear trend. Note that the age and income groups were numerically mapped for easier visualization.</p>

<hr>

<h3 style="color: #2E86C1; font-family: Arial, sans-serif;">üìà <strong>Gender vs Star Wars Film Fandom</strong></h3>
<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">Among males and females, the majority consider themselves to be fans of the Star Wars franchise. However, the number of male fans appears to be significantly higher than female fans.</p>

<hr>

<h3 style="color: #2E86C1; font-family: Arial, sans-serif;">üìà <strong>Age vs Star Wars Film Fandom</strong></h3>
<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">Across various age groups, the majority consider themselves fans of the Star Wars franchise. The age group "45-60" has the highest number of fans, followed by "30-44".</p>

<hr>

<h3 style="color: #34495E; font-family: Arial, sans-serif;">üìä <strong>4.3 Multivariate Analysis</strong></h3>
<ul style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">
    <li>Pairplot or heatmap to show correlations between multiple variables, if applicable</li>
</ul>



```python
# Multivariate Analysis

# Select only the columns that are relevant for the multivariate analysis
# For demonstration purposes, I've selected gender, age, household income, and fandom status
selected_columns = ['Gender', 'Age_numeric', 'Income_numeric', 'Do you consider yourself to be a fan of the Star Wars film franchise?']
df_selected = df[selected_columns].dropna()

# Map 'Yes' and 'No' in fandom status to 1 and 0 respectively for easier correlation analysis
df_selected['Fandom_numeric'] = df_selected['Do you consider yourself to be a fan of the Star Wars film franchise?'].map({'Yes': 1, 'No': 0})

# Map 'Male' and 'Female' in gender to 1 and 0 respectively for easier correlation analysis
df_selected['Gender_numeric'] = df_selected['Gender'].map({'Male': 1, 'Female': 0})

# Calculate the correlation matrix
correlation_matrix = df_selected[['Gender_numeric', 'Age_numeric', 'Income_numeric', 'Fandom_numeric']].corr()

# Draw the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Selected Variables")
plt.show()

# Pairplot to visualize relationships between numerical variables
sns.pairplot(df_selected, hue='Do you consider yourself to be a fan of the Star Wars film franchise?', diag_kind='kde')
plt.suptitle('Pairplot of Selected Variables', y=1.02)
plt.show()
```

<h3 style="color: #2E86C1; font-family: Arial, sans-serif;">üìà <strong>Correlation Matrix of Selected Variables</strong></h3>
<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">The correlation matrix shows that there is no strong correlation between the selected variables (Gender, Age, Household Income, and Star Wars Film Fandom). The highest correlation value is approximately 0.16, indicating a very weak correlation.</p>

<hr>

<h3 style="color: #2E86C1; font-family: Arial, sans-serif;">üìà <strong>Pairplot of Selected Variables</strong></h3>
<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">
    The pairplot provides a visualization of the relationships between the selected numerical variables. Different colors represent whether the respondent is a fan of the Star Wars franchise or not. The KDE (Kernel Density Estimation) plots along the diagonal show the distribution of each variable. No clear patterns or trends are visible in the scatter plots, reinforcing the findings from the correlation matrix that there are no strong correlations between these variables.
</p>

<h2 style="color: #34495E; font-family: Arial, sans-serif;">üîç <strong>Extra Analysis: Advanced Techniques</strong></h2>

<h3 style="color: #2E86C1; font-family: Arial, sans-serif;">üìä <strong>Clustering to Identify Distinct Groups</strong></h3>
<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">
  In order to unravel the underlying structure of our dataset, we'll employ the K-means clustering technique. This is an unsupervised machine learning algorithm that could provide insights into the heterogeneity of the respondents.
</p>

<h3 style="color: #2E86C1; font-family: Arial, sans-serif;">üìä <strong>Data Preprocessing for Clustering</strong></h3>
<ol style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 40px;">
  <li><strong>Select Relevant Variables:</strong> The first step involves picking a subset of variables that are likely to reveal distinct clusters. Given that K-means is sensitive to the scale of the variables, the selection should ideally involve numerical variables.</li>
  <li><strong>Preprocess the Data:</strong> Before applying the K-means algorithm, we'll need to preprocess the data. This involves handling missing values, scaling the variables, and encoding categorical variables.</li>
  <li><strong>Apply K-means and Visualize:</strong> After preprocessing, we'll apply the K-means algorithm to identify clusters. We'll then visualize these clusters to interpret their characteristics.</li>
</ol>

<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">Let's start by selecting and preprocessing the data.</p>

<hr>


```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Selecting a subset of columns for clustering
selected_columns = ['Gender', 'Age', 'Household Income', 'Do you consider yourself to be a fan of the Star Wars film franchise?']
df_clustering = df[selected_columns].dropna()

# Encoding categorical variables
label_encoder = LabelEncoder()
for col in selected_columns:
    df_clustering[col] = label_encoder.fit_transform(df_clustering[col])

# Standardizing the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_clustering)

# Applying PCA for dimensionality reduction
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_scaled)
principal_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])

# Finding the optimal number of clusters using the elbow method
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the elbow graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method For Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

<hr>

<h3 style="color: #2E86C1; font-family: Arial, sans-serif;">üìà <strong>Determining the Optimal Number of Clusters</strong></h3>
<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">
  - The elbow method is commonly used to find the optimal number of clusters for K-means clustering. The point where the inertia starts to decrease at a slower rate ("the elbow") is generally considered the optimal number of clusters.
  - In this case, the elbow appears to be around 2 or 3 clusters.
</p>

<p style="font-size: 16px; font-family: Arial, sans-serif; margin-left: 20px;">
  Next, I'll proceed with applying K-means clustering using 3 clusters and then visualize these clusters in the 2D space formed by the first two principal components.
</p>

<hr>


```python
#Dropping rows where gender or age is NaN for the purpose of visualization
df = df.dropna(subset=['Gender', 'Age'])

# Selecting a subset of columns for clustering
selected_columns = ['Gender', 'Age', 'Household Income', 'Do you consider yourself to be a fan of the Star Wars film franchise?']
df_clustering = df[selected_columns].dropna()

# Encoding categorical variables
label_encoder = LabelEncoder()
for col in selected_columns:
    df_clustering[col] = label_encoder.fit_transform(df_clustering[col])

# Standardizing the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_clustering)

# Applying PCA for dimensionality reduction
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_scaled)
principal_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])

# Applying K-means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(df_scaled)

# Adding the cluster labels to the principal component dataframe
principal_df['Cluster'] = clusters

# Visualizing the clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Principal Component 1', y='Principal Component 2', hue='Cluster', data=principal_df, palette='viridis')
plt.title('Clusters of Respondents')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
